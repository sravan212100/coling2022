# 5 fold

```python
merge_df_a = pd.concat([train_twitter,valid_twitter])
merge_df_b = pd.concat([train_web,valid_web])
merge_df_b
```

```python
merge_df_a.to_csv('merge_a.tsv', sep='\t')
merge_df_b.to_csv('merge_b.tsv', sep='\t')
```

```python
merge_df_a = pd.read_csv('merge_a.tsv', sep='\t')
merge_df_a = merge_df_a.drop('Unnamed: 0',axis=1)
merge_df_b = pd.read_csv('merge_b.tsv', sep='\t')
merge_df_b = merge_df_b.drop('Unnamed: 0',axis=1)
merge_df_a
```

```python
size = len(merge_df_a.label)
for i in range(5):
    merge_df_a[int(i*0.2*size):int((i+1)*0.2*size)].to_csv('validation_a_{}.tsv'.format(i), sep='\t')
    merge_df_a.drop(list(range(int(i*0.2*size),int((i+1)*0.2*size)))).to_csv('train_a_{}.tsv'.format(i), sep='\t')
    
size = len(merge_df_b.label)
for i in range(5):
    merge_df_b[int(i*0.2*size):int((i+1)*0.2*size)].to_csv('validation_b_{}.tsv'.format(i), sep='\t')
    merge_df_b.drop(list(range(int(i*0.2*size),int((i+1)*0.2*size)))).to_csv('train_b_{}.tsv'.format(i), sep='\t')
```
